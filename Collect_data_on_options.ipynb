```python
from yahoo_fin import options as op
import pandas as pd
import numpy as np
```

```python
companies = ['AAPL', 
            'GOOGL',
            'MSFT',
            'SPY',
            'PFE',
            'LMT',
            'DIS',
            'VZ',
            'PYPL',
            'TSN',
            'DG',
            'QCOM',
            'META',
            'TSLA',
            'NFLX',
            'NVDA',
            'AMZN',
            'AVGO',
            'LLY',
            'JPM',
            'UNH',
            'V',
            'MA',
            'JNJ',
            'PG',
            'MRK',
            'MCD',
            'WMT',
            'BAC',
            'ACN',
            'KO',
            'PEP',
            'CSCO',
            'ORCL',
            'INTC']
```

```python
# 34 companies and S&P 500
len(companies) - 1
```

```python
# put down the date, so that we can save file and understand at what date the data was collected
current_date = '12 April'
dataset_all = pd.DataFrame()
```

```python
# a script to collect data
for TICKER in companies:
    print(TICKER)
    ticker_data = pd.DataFrame()
    try:
        all_dates = op.get_expiration_dates(TICKER)
        for date in all_dates:
            df = op.get_calls(TICKER, date=date)
            df['Expiration date'] = date
            df['Company'] = TICKER
            df['Date of getting data'] = current_date
            ticker_data = ticker_data.append(df)

        ticker_data.index = np.arange(0, len(ticker_data))

        dataset_all = dataset_all.append(ticker_data)
        print('Success')
    except Exception as e:
        print('Error with this TICKER')
        print(str(e))
        pass
    #time.sleep(5)
```

```python
# save
dataset_all.to_csv(f'DATASETS/NEW_{current_date}.csv', index=False)
```
