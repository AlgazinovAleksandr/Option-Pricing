# Goal of the notebook

Remember that I had to collect data every day to make it historical (because in Yahoo one cannot get historical data on options). Hence, this notebook is required to put all data into a single file.

```python
import pandas as pd
import numpy as np
```

```python
# put down all the files with data
paths = ['DATASETS/NEW_16 February.csv', 'DATASETS/NEW_19 February.csv', 'DATASETS/NEW_20 February.csv',
        'DATASETS/NEW_21 February.csv', 'DATASETS/NEW_22 February.csv', 'DATASETS/NEW_23 February.csv',
        'DATASETS/NEW_26 February.csv', 'DATASETS/NEW_27 February.csv', 'DATASETS/NEW_28 February.csv',
        'DATASETS/NEW_29 February.csv', 'DATASETS/NEW_1 March.csv', 'DATASETS/NEW_2 March.csv', 
        'DATASETS/NEW_4 March.csv', 'DATASETS/NEW_5 March.csv', 'DATASETS/NEW_6 March.csv',
        'DATASETS/NEW_7 March.csv', 'DATASETS/NEW_8 March.csv',
        'DATASETS/NEW_11 March.csv', 'DATASETS/NEW_12 March.csv', 'DATASETS/NEW_13 March.csv',
        'DATASETS/NEW_14 March.csv', 'DATASETS/NEW_15 March.csv',
        'DATASETS/NEW_18 March.csv', 'DATASETS/NEW_19 March.csv', 'DATASETS/NEW_20 March.csv',
        'DATASETS/NEW_21 March.csv', 'DATASETS/NEW_22 March.csv',
        'DATASETS/NEW_25 March.csv', 'DATASETS/NEW_26 March.csv', 'DATASETS/NEW_27 March.csv',
         'DATASETS/NEW_28 March.csv', 'DATASETS/NEW_29 March.csv',
        'DATASETS/NEW_1 April.csv', 'DATASETS/NEW_2 April.csv', 'DATASETS/NEW_3 April.csv', 'DATASETS/NEW_4 April.csv',
        'DATASETS/NEW_8 April.csv', 'DATASETS/NEW_9 April.csv', 'DATASETS/NEW_10 April.csv',
        'DATASETS/NEW_11 April.csv', 'DATASETS/NEW_12 April.csv']
```

```python
# a function that takes all file paths and constructs a single file
def get_data(datasets_paths):
    dataset_full = pd.DataFrame()
    for path in datasets_paths:
        df = pd.read_csv(path)
        dataset_full = dataset_full.append(df)
    dataset_full.index = np.arange(0, len(dataset_full))
    return dataset_full
```

```python
dataset_full = get_data(paths)
```

```python
# drop the unnecessary column
dataset_full = dataset_full.drop('0', axis=1)
dataset_full
```

```python
dataset_full = dataset_full.drop_duplicates()
dataset_full
```

```python
# save data
dataset_full.to_csv('Raw_data_02_05.csv', index=None)
```
